```python
import esmvalcore.esgf
import numpy as np
```

```python
for ESM in ['ACCESS-CM2', 'BCC-CSM2-MR', 'CESM2', 'CESM2-WACCM', 'CNRMCM6-1', 'CNRM-CM6-1-HR', 'CNRM-ESM2-1', 'EC-Earth3', 'EC-Earth3-Veg', 'GFDL-CM4', 'GFDL-ESM4', 'HadGEM3GC31-LL', 'HadGEM-GC31-MM', 'MPI-ESM1-2-HR', 'MRI-ESM2-0', 'KACE-1-0-G', 'TaiEMS1', 'UKESM1-0LL']:
    fls = esmvalcore.esgf.find_files(
        project='CMIP6',
        mip='day',
        short_name='tas',
        dataset=ESM,
        exp='ssp370',
        ensemble='*',
    )  
    ensemble_names = np.unique([fl.__dict__['facets']['ensemble'] for fl in fls])

    for ensemble in ensemble_names:
        download = True
        files_to_download = []
        for var in ['tas','tasmax','rsds','sfcWind']:
            fls = esmvalcore.esgf.find_files(
                project='CMIP6',
                mip='day',
                short_name=var,
                dataset=ESM,
                exp='ssp370',
                ensemble=ensemble,
            )  
            files_to_download = []
            for fl in fls:
                files_to_download.append(fl)
                if int(fl.name.split('-')[-1][:4]) >= 2024:
                    break
            #files_to_download += [fl for fl in fls if '20150101' in fl.name]
            if len([fl for fl in fls if '20150101' in fl.name]) == 0:
                download = False

        if download:
            print(ESM,files_to_download)
            esmvalcore.esgf.download(files_to_download, '/climca/data/', n_jobs=4)
```
